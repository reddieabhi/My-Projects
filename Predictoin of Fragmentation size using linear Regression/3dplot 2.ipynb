```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from mpl_toolkits import mplot3d
from mpl_toolkits.mplot3d import Axes3D
ax = plt.axes(projection = '3d')
from matplotlib import colorbar



data = pd.read_csv("C:\\Users\\babhi\\Downloads\\read2.csv")
data = data.iloc[:,:9]
data.columns = ['B','StoB','HtoB','BtoD','TtoB','PF','Xb','E','D50']
x = np.array(data['BtoD'])
y = np.array(data['StoB'])


X,Y = np.meshgrid(x,y)


z = 0.0030360930059659097*X+0.042049090904264085*Y
ax.plot_surface(X,Y,z,cmap = 'rainbow')
plt.contour(X,Y,z,cmap='rainbow')
ax.contour(X,Y,z)
plt.colorbar()
plt.show()

```
