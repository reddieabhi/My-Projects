```python
x= np.linspace(-3,3,100)
y = np.linspace(-3,4,100)
z = lambda x,y: np.sin(x)*np.cos(y)
X,Y = np.meshgrid(x,y)
F = z(X,Y)
```

```python
ax.plot_surface(X,Y,F)
```

```python
plt.show()
```

```python
plt.contour(X,Y,F,cmap='rainbow')
plt.colorbar()
plt.show()
```
