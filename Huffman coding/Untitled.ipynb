```python
class Binary_tree():
    def __init__(self,val,prio):
        self.val = val
        self.prio = prio
        self.left = None
        self.right = None
    
```

```python
class priority_queue_node():
    def __init__(self,val,prio):
        self.val = val
        self.prio = prio
        

class min_priority_queue():
    def __init__(self):
        self.arr = []
        self.size = 0
        
    def size_of_heap(self):
        return self.size
    
    
        
    def percolate_up(self):
        child = self.size - 1
        while child > 0:
            parent = (child-1)//2
            if self.arr[parent].prio <= self.arr[child].prio:
                break
            else:
                self.arr[parent],self.arr[child] = self.arr[child],self.arr[parent]
                child = parent
        return 
    
    
    def percolate_down(self):
        parent = 0
        n = self.size
        c1 = 2*parent + 1
        c2 = 2*parent + 2
        if c2 == n:
            if self.arr[parent].prio <= self.arr[c1].prio:
                return
            else:
                self.arr[parent],self.arr[c1] = self.arr[c1],self.arr[parent]
                return
                
        while parent < n:
            c1 = 2*parent + 1
            c2 = 2*parent + 2
            if c1 >= n:
                break
            if c2 == n:
                if self.arr[parent].prio <= self.arr[c1].prio:
                    return
                else:
                    self.arr[parent],self.arr[c1] = self.arr[c1],self.arr[parent]
                    return
            
            
            
            min_prio = c1
            if self.arr[c2].prio < self.arr[c1].prio:
                min_prio = c2
            if self.arr[parent].prio <= self.arr[min_prio].prio:
                return
            self.arr[parent],self.arr[min_prio] = self.arr[min_prio],self.arr[parent]
            parent = min_prio
            
        return
    
    
    def insert(self,node):
        self.arr.append(node)
        self.size += 1
        self.percolate_up()
        
        
            
    def give_minimum(self):
        if self.size <= 0:
            return 'heap is empty'
        n = self.size
        self.arr[0],self.arr[n-1] = self.arr[n-1],self.arr[0]
        ans_node = self.arr.pop()
        self.size -= 1
        self.percolate_down()
        return ans_node
       
    
    
        
    
```
