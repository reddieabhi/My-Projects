```python
def gradient_descent(x,y):
    n = len(x)
    mcurr = 0
    ccurr = 0
    inter = 1000
    learning_gradient = 0.01
    for i in range(inter):
        
        ypred = mcurr*x + ccurr
        cost = (1/n)*(sum([val**2 for val in (ypred - y)]))
        mderiv = -(2/n)*sum(x*(y-ypred))
        cderiv = -(2/n)*sum((y-ypred))
        mcurr -= learning_gradient*mderiv
        ccurr -= learning_gradient*cderiv
        print(f'm-{mderiv},c-{cderiv},cost-error {cost} iter-{i}')
        
```

```python
gradient_descent(a,b)
```
